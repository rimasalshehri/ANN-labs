```python
def logsiglayer(W1, p, b1) :
  n=(np.dot(W1, p))+b1
  a1=logsigmoid(n)
  return a1

def linearlayer (W2, a1, b2) :
   n=(np.dot(W2, a1))+b2
   a2=purelin(n)
   return a2
```

```python
def updateparams(W1, W2, b1, b2, s1, s2, a0, a1, a2, alpha) :
 W1 = W1- (alpha *s1)*np.transpose(a0)
 b1= b1-alpha*s1
 W2= W2- alpha*(np.dot(s2, a1.T))
 b2= b2-alpha*s2
 return W1, b1, W2, b2
```

```python
def backpropagatealgorithm(W1, W2, b1, b2, p, t, alpha):
 a0, a1, a2 = forwardpropagation(p, W1, b1, W2, b2)
 f1= dlogsig(a1)
 F1= np.diag(f1)
 F2= dpurelin(a2)
 s1, s2= backpropagation(F2, t, a2, F1, W2)
 W1, b1, W2, b2= updateparams(W1, W2, b1, b2, s1, s2, a0, a1, a2, alpha)
 print("W1= ", W1)
 print("b1= ", b1)
 print("W2= ", W2)
 print("b2= ", b2)
```
