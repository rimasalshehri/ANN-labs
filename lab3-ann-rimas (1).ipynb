# Task 3 & 4

```python
import numpy as np

def recurrentlayer(p, w, t):
    if t == 0:
        a2 = p
    else:
        a2 = np.maximum(0, np.dot(w, p))
    return a2

# Given values
a1 = np.array([4, 2])
W = np.array([[1, -0.5], [-0.5, 1]])

# Table
table = []
for t in range(3):
    a2 = recurrentlayer(a1, W, t)
    table.append((t, a1, a2))
    a1 = a2  # Update a1 for the next timestep

# Print the table
print("t\t\tInput\t\t\tOutput")
for t, a1, a2 in table:
    print(f"{t}\t\t{a1}\t\t{a2}")
```

```python
import numpy as np

def feedforwardlayer(p, w, b):
    n = np.dot(w, p) + b
    return n

def poslin(x):
    return np.maximum(0, x)

def recurrentlayer(p, w, t):
    a2 = p  # Initialize a2 with p
    if t > 0:
        for _ in range(t):
            a2 = poslin(np.dot(w, a2))
    return a2

# Given parameters and input values
W1 = np.array([[1, -1, -1], [1, 1, -1]])
b1 = np.array([3, 3])
W2 = np.array([[1, -0.5], [-0.5, 1]])
p = np.array([-1, 1, -1])

# Feedforward layer
a1 = feedforwardlayer(p, W1, b1)

# Recurrent layer with t=1
t = 1
a2 = recurrentlayer(a1, W2, t)
print(a2)

# Recurrent layer with t=2
t = 2
a2 = recurrentlayer(a1, W2, t)
print(a2)
```
