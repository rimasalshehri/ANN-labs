```python
import matplotlib.pyplot as plt #here we import matplotlib library to do the plotting and visualization
import numpy as np #here we import numpy so that we can deal with mathematical equations and numbers
```

Q1

```python
#here we implement the function of logsigmoid by allowing the function to recieve a value and then returning the 
#solution of the equation 
def sigmoid(z):   
  return 1.0 / (1 + np.exp(-z))
```

```python
#the values we plot for are -5 and 5
n = np.arange(-5, 5)
x = sigmoid(n)
```

```python
#here we do the ploting and supply the labels of x and y axis
plt.figure(figsize=(3, 3))
plt.plot(n,x )
plt.xlabel('n')
plt.ylabel('Output')
plt.title('Sigmid Transfer Function')
plt.grid(True)
plt.show()
```

```python
#here we implement the function of hardlimit by allowing the function to recieve a value and then returning the 
#solution of the equation 
def hardlim(z):
   return np.where(z > 0, 1 , 0)
```

```python
#the values we plot for are -5 and 5
n = np.arange(-5, 5)
x = hardlim(n)
```

```python
#here we do the ploting and supply the labels of x and y axis
plt.figure(figsize=(3, 3))
plt.plot(n,x)
plt.xlabel('n')
plt.ylabel('Output')
plt.title('Hardlimit Transfer Function')
plt.grid(True)
plt.show()
```

```python
#here we implement the function of hyperbolic tangent by allowing the function to recieve a value and then returning the 
#solution of the equation 
def tanh(z):
  return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
```

```python
#the values we plot for are -5 and 5
n = np.arange(-5, 5)
x = tanh(n)
```

```python
#here we do the ploting and supply the labels of x and y axis
plt.figure(figsize=(3, 3))
plt.plot(n,x )
plt.xlabel('n')
plt.ylabel('Output')
plt.title('tanh Transfer Function')
plt.grid(True)
plt.show()
```
