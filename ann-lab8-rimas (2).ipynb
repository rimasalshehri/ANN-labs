# Q2

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Set random state for make_classification
X, y = make_classification(n_samples=100, random_state=42)

# Set random state for train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

# Set random state for MLPClassifier
clf = MLPClassifier(max_iter=10, random_state=42).fit(X_train, y_train)

# Report the score
score = clf.score(X_test, y_test)
print("Run 1: Score:", score)

# Repeat the example for five times
for run in range(2, 6):
    # Set random state for make_classification
    X, y = make_classification(n_samples=100, random_state=run)

    # Set random state for train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=run)

    # Set random state for MLPClassifier
    clf = MLPClassifier(max_iter=10, random_state=run).fit(X_train, y_train)

    # Report the score
    score = clf.score(X_test, y_test)
    print(f"Run {run}: Score: {score}")
```

# Q3

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Function to run the example with different max_iter values
def run_example(max_iter_value):
    X, y = make_classification(n_samples=100)
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)

    clf = MLPClassifier(max_iter=max_iter_value).fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    return score

# Table headers
print("Max_iter\tScore")

# Run the example for different max_iter values
max_iter_values = [50, 100, 200, 300]
for max_iter in max_iter_values:
    score = run_example(max_iter)
    print(f"{max_iter}\t\t{score}")
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# 1. Split the dataset into 70% training and 30% testing
X, y = make_classification(n_samples=100)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Initialize an MLPClassifier with 1 hidden layer and 7 neurons
clf = MLPClassifier(hidden_layer_sizes=(7,), learning_rate_init=0.1, max_iter=1000)

# 3. Fit the model given the training set
clf.fit(X_train, y_train)

# 4. Predict the output of the testing set
y_pred = clf.predict(X_test)

# 5. Compute and report the score
score = clf.score(X_test, y_test)
print("Score:", score)

# 7. What is the effect of the number of neurons on the result?
# Repeat steps 2 to 6, but change the number of neurons in the hidden layer
neurons = [3, 5, 10]
for n in neurons:
    clf = MLPClassifier(hidden_layer_sizes=(n,), learning_rate_init=0.1, max_iter=1000)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print("Number of neurons:", n)
    print("Score:", score)

# 8. What is the effect of the number of hidden layers on the result?
# Repeat steps 2 to 6, but change the number of hidden layers
hidden_layers = [1, 2, 3]
for n_layers in hidden_layers:
    clf = MLPClassifier(hidden_layer_sizes=(7,) * n_layers, learning_rate_init=0.1, max_iter=1000)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print("Number of hidden layers:", n_layers)
    print("Score:", score)

# 9. Apply grid search (GridSearchCV) to find the best hyperparameters
from sklearn.model_selection import GridSearchCV

param_grid = {
    'hidden_layer_sizes': [(7,), (7, 7), (7, 7, 7)],
    'learning_rate_init': [0.01, 0.1, 1.0],
    'max_iter': [500, 1000, 1500]
}

clf = MLPClassifier()
grid_search = GridSearchCV(clf, param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Hyperparameters:", best_params)
print("Best Score:", best_score)
```
