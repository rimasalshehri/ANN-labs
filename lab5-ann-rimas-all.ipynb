# Q1 Linear associator network

```python
def purelin(n):
    return n  # Identity function

def linear_associator(W, p):
    a = np.dot(W, p)  # Matrix multiplication between W and p
    a = purelin(a)  # Apply the purelin transfer function
    return a

# Example usage of Q1:
W = np.array([[0.5, 0.8, 0.2], [0.3, 0.4, 0.9]])  # Weight matrix
p = np.array([2, 3, 1])  # Input vector

output = linear_associator(W, p)
print(output)
```

# Q2 Hebb learning rule

```python
def hebb_rule(T,p):
    p = np.transpose(p)
    w = np.dot(T,p)
    return w
```

# Testing Hebb rule

```python
p = [[0.5,0.5], [-0.5,0.5], [0.5,-0.5], [-0.5,-0.5]]
t = [[1,1], [-1, 1]]
w = hebb_rule(t,p)
print (w)
```

# Q3 Function that checks if two vectors are orthogonal

```python
def check_orthogonal(a,b):
    b = np.transpose(b)
    c = np.dot(a,b)
    return c
```

# Testing the function

```python
p1 = [0.5,-0.5,0.5,-0.5]
p2 = [0.5,0.5,-0.5,-0.5]
c = check_orthogonal(p1,p2)
print(c)
```

# Q4 a function that checks if a vector is of unit length

```python
def check_unitlength(a):
    a_transpose = np.transpose(a)
    c = np.dot(a,a_transpose)
    return c
```

# Test the function ‚Äúcheck_unitlength‚Äù 

```python
p = [0.5,-0.5,0.5,-0.5]
c = check_unitlength(p)
print(c)
```

```python
def normalizevec(a):
    a = np.array(a)
    norm = np.linalg.norm(a)
    c = a / norm
    return c
p1 = [1, -1, -1]
print(normalizevec(p1))
```
